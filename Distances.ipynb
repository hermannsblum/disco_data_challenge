```python
cats = {}
for idx, restaurant in restaurants.iterrows():
    neighbours = cells.get_neighbours(restaurant)
    index = pd.DataFrame(neighbours)
    index = index.sort('distance', ascending=True)[1:6]['index'].tolist()
    for idx, neighbour in businesses.loc[index].iterrows():
        for category in neighbour.categories:
            cats[category] = cats.get(category, 0) + 1
```

```python
# 10 biggest categories
ten_categories = [
    'Shopping',
     'Restaurants',
     'Food',
     'Event Planning & Services',
     'Active Life',
     'Arts & Entertainment',
     'Local Services',
     'Nightlife',
     'Home Services',
     'Health & Medical',
     'Hotels & Travel',
     'Bars',
     'Specialty Food',
     'Education',
     'Cafes',
     'Fashion'
]

# now the big work: Map Categories in Neighbourhoods together
df = pd.DataFrame()
for category in ten_categories:
    is_category = businesses.apply(lambda row: category in row.categories, axis=1)
    in_category = businesses[is_category == 1]
    cats = {} # dict of category-mappings for items of category
    for idx, in_item in in_category.iterrows():
        neighbours = cells.get_neighbours(in_item)
        index = pd.DataFrame(neighbours)
        # get indizes of first 5 neighbours
        index = index.sort('distance', ascending=True)[:5]['index'].tolist()
        for idx, neighbour in businesses.loc[index].iterrows():
            for cat in neighbour.categories:
                # we simply count the occurences
                cats[cat] = cats.get(cat, 0) + 1
    if category in list(df.columns.values):
        df[category] = df[category] + pd.Series(cats)
    else:
        df[category] = pd.Series(cats)
```

```python
not_mentioned = []
for category in categories:
    if category not in list(df.columns.values):
        not_mentioned.append(category)
print("{} out of {}".format(len(not_mentioned), len(categories)))
```
