```python
# now the big work: Map Super-Categories in Neighbourhoods together
cat_map = pd.DataFrame()
for supercat in list(names.keys()):
    in_category = businesses[businesses['super category'] == supercat]
    cats = {} # dict of category-mappings for items of category
    for idx, in_item in in_category.iterrows():
        neighbours = cells.get_neighbours(in_item, num=1)
        index = pd.DataFrame(neighbours)
        # get indizes of first 5 neighbours
        index = index.sort('distance', ascending=True)[:5]['index'].tolist()
        for idx, neighbour in businesses.loc[index].iterrows():
            neighbour_cat = neighbour['super category']
            # we simply count the occurences
            cats[neighbour_cat] = cats.get(neighbour_cat, 0) + 1
    # add to the map of categories
    cat_name = names[supercat]
    if cat_name in list(cat_map.columns.values):
        cat_map[cat_name] = cat_map[cat_name] + pd.Series(cats)
    else:
        cat_map[cat_name] = pd.Series(cats)
```
